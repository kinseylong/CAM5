```python
import pandas as pd
```

```python
calmodulin_proteins = pd.read_csv("data/TAIR10/CalmodulinProteinsUniprot.tsv", sep="\t")
calmodulin_proteins.head()
```

```python
calmodulin_proteins[calmodulin_proteins["Gene Names"].str.contains("calmodulin", na=False)]
```

```python
filtered = calmodulin_proteins[calmodulin_proteins["Gene Names"].str.contains(r"\bCAM\d+\b", na=False)]

filtered = filtered[["Protein names", "Gene Names (ordered locus)", "Sequence"]].drop_duplicates()

with open("calmodulin_sequences.fasta", "w") as f:
    for i, row in filtered.iterrows():
        f.write(f">{i} {row['Protein names']}{row['Gene Names (ordered locus)']}  \n{row['Sequence']}\n")
        
```

```python
from collections import defaultdict

fasta_counts = defaultdict(int)

with open("filtered_calmodulin_sequences.fasta", "w") as f:
    filtered_dash = filtered.copy()
    filtered_dash["Protein names"] = filtered_dash["Protein names"].str.replace(" ", "-", regex=False)
    for _, row in filtered_dash.iterrows():
        name = row['Protein names']
        fasta_counts[name] += 1
        if fasta_counts[name] > 1:
            fasta_id = f"{name}_{fasta_counts[name]}"
        else:
            fasta_id = name
        f.write(f">{fasta_id}\n{row['Sequence']}\n")
```
